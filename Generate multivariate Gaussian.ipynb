```
import numpy as np
import matplotlib.pyplot as plt
import scipy.linalg as spla
from mpl_toolkits.mplot3d import axes3d
```

##a)

```
def GSAMPLE(mu, Sigma, n):
    L = spla.cholesky(Sigma) #Here, using cholesky decomposition method
    Z = np.random.normal(size=(n, Sigma.shape[0]))
    G = Z.dot(L) + mu
    return (G)
```

##b)

```
# Generate a large sample of n=1000 by GSAMPLE:
mean = np.array([1., -5. ])
covariance = np.array([[1., 0.3], [0.3, 2.]])
X = GSAMPLE(mean, covariance, n = 1000)
```

```
#Verify GSAMPLE works by examing its mean:
np.mean(X, axis = 0)
```

```
#Verify GSAMPLE works by examing its covariance:
np.cov(X.T)
```

##c)

```
n = 100

#For two dimemsions: 
mu = np.array([1., 1.])
sigma = np.array([[2., 1.], [ 1., 2.]])
X = GSAMPLE(mu, sigma, n)

plt.scatter(X.T[0,:], X.T[1,:], s = 30) 
#Generated by multiplying normally distributed points by the Cholesky decomposition matrix
plt.show() 
```

```
#For three dimensions:
n = 100
mu = np.array([1., 1., 1.])
sigma = np.array([[5., 2., 0], [2., 3., 1.], [0, 1., 1.]])
Y = GSAMPLE(mu, sigma, n)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(Y[:,0], Y[:,1], Y[:,2])
ax.set_xlabel('X label')
ax.set_ylabel('Y label')
ax.set_zlabel('Z label')
plt.show()
```

According to above histogram plots, we could draw the following conclusions:

2). For fixed large sample size, larger bins tend to approximate distribution that closer to true distribution then small bins, and the plot will also appear much more smoothly;

3). For small sample size with medium to large bins, the estimated distribution will not be accurate or reliable.
